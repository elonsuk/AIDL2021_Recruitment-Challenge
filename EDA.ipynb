```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('Data_Files/train.csv')
data.head()
```

```python
language = pd.get_dummies(data['Language'], drop_first=True)
language.head()
```

```python
data = pd.concat([data, language], axis=1)
data.head()
```

```python
data.drop('Language', axis=1, inplace=True)
data.head()
```

```python
data.drop('Channel', axis=1, inplace=True)
data.head()
```

```python
numeric = data[['Watch time(Minutes)', 'Stream time(minutes)', 'Peak viewers', 'Average viewers', 'Followers', 
                'Followers gained', 'Views gained']]
numeric.head()
```

```python
numeric['Wt/St'] = numeric['Watch time(Minutes)'] / numeric['Stream time(minutes)']
numeric['Wt/Pv'] = numeric['Watch time(Minutes)'] / numeric['Peak viewers']
numeric['Wt/Av'] = numeric['Watch time(Minutes)'] / numeric['Average viewers']
numeric['Wt/F'] = numeric['Watch time(Minutes)'] / numeric['Followers']
numeric['Wf/Vg'] = numeric['Watch time(Minutes)'] / numeric['Views gained']

numeric['St/Pv'] = numeric['Stream time(minutes)'] / numeric['Peak viewers']
numeric['St/Av'] = numeric['Stream time(minutes)'] / numeric['Average viewers']
numeric['St/F'] = numeric['Stream time(minutes)'] / numeric['Followers']
numeric['St/Vg'] = numeric['Stream time(minutes)'] / numeric['Views gained']

numeric['Pv/Av'] = numeric['Peak viewers'] / numeric['Average viewers']
numeric['Pv/F'] = numeric['Peak viewers'] / numeric['Followers']
numeric['Pv/Vg'] = numeric['Peak viewers'] / numeric['Views gained']

numeric['Av/F'] = numeric['Average viewers'] / numeric['Followers']
numeric['Av/Vg'] = numeric['Average viewers'] / numeric['Views gained']

numeric['F/Vg'] = numeric['Followers'] / numeric['Views gained']

numeric.head()
```

```python
extracted = numeric.corr().loc['Followers gained']
extracted = pd.DataFrame(extracted).transpose()
extracted.drop('Followers gained', axis=1, inplace=True)

figure, ax = plt.subplots(figsize=(24,4))
sns.heatmap(ax=ax, data=extracted, annot=True)
```

```python
sig_elements = numeric[['Watch time(Minutes)', 'Peak viewers', 'Average viewers', 'Followers', 'Views gained', 
                     'Wt/St', 'F/Vg']]
sig_elements.head()
```

```python
extracted = extracted[sig_elements.columns]

figure, ax = plt.subplots(figsize=(24,4))
sns.heatmap(ax=ax, data=extracted, annot=True)
```

```python
extracted.drop(['Views gained', 'F/Vg'], axis=1, inplace=True)

figure, ax = plt.subplots(figsize=(24,4))
sns.heatmap(ax=ax, data=extracted, annot=True)
```

```python
final_data = numeric[extracted.columns]
final_data.head()
```

```python
X = final_data
y = data['Followers gained']
```
