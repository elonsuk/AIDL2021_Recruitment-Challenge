```python
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
import keras.backend as K
from keras.losses import mean_squared_error as mse
```

## Load & Process Test Data

```python
test = pd.read_csv('Data_Files/test.csv')

test.drop(['Channel', 'Partnered', 'Mature'], axis=1, inplace=True)
test.rename(columns={'Watch time(Minutes)': 'Watch time', 'Stream time(minutes)': 'Stream time'}, inplace=True)
        
ENG = []
KOR = []
RUS = []
OTH = []

for i in range(len(test)):
    if (test.iloc[i]['Language'] == 'English'):
        ENG.append(1)
        KOR.append(0)
        RUS.append(0)
        OTH.append(0)
        
    elif (test.iloc[i]['Language'] == 'Korean'):
        KOR.append(1)
        ENG.append(0)
        RUS.append(0)
        OTH.append(0)
        
    elif (test.iloc[i]['Language'] == 'Russian'):
        RUS.append(1)
        ENG.append(0)
        KOR.append(0)
        OTH.append(0)
    
    else:
        OTH.append(1)
        ENG.append(0)
        KOR.append(0)
        RUS.append(0)

test['English'] = ENG
test['Korean'] = KOR
test['Russian'] = RUS
test['Others'] = OTH

test.drop('Language', axis=1, inplace=True)

test['VS'] = test['Views gained'] / test['Stream time']

columns = ['Watch time', 'Followers', 
           'Peak viewers', 'Average viewers', 
           'English', 'Korean', 'Russian', 'Others',
           'VS']
test = test.reindex(columns=columns)

test = scaler.fit_transform(test.values)
```

## Model Generation, Training, and Prediction

```python
def create_model():

    inputs = Input(shape=(X_train.shape[-1]))

    x1 = Dense(430, activation='selu', kernel_initializer=keras.initializers.RandomUniform())(inputs)
    d1 = Dropout(0.9)(x1)

    x2 = Dense(430, activation='selu', kernel_initializer=keras.initializers.RandomUniform())(d1)
    d2 = Dropout(0.8)(x2)

    x3 = Dense(256, activation='selu', kernel_initializer=keras.initializers.RandomUniform())(d2)
    d4 = Dropout(0.7)(x3)

    x4 = Dense(256, activation='selu', kernel_initializer=keras.initializers.RandomUniform())(d4)
    d5 = Dropout(0.6)(x4)

    x5 = Dense(128, activation='selu', kernel_initializer=keras.initializers.RandomUniform())(d5)
    d6 = Dropout(0.7)(x5)

    x6 = Dense(128, activation='selu', kernel_initializer=keras.initializers.RandomUniform())(d6)
    d7 = Dropout(0.9)(x6)

    x7 = Dense(32, activation='selu', kernel_initializer=keras.initializers.RandomUniform())(d7)
    d8 = Dropout(0.9)(x7)

    x8 = Dense(32, activation='selu', kernel_initializer=keras.initializers.RandomUniform())(d8)
    d9 = Dropout(0.9)(x8)

    outputs = Dense(1)(d9)

    model = keras.Model(inputs=inputs, outputs=outputs)
    
    return model
```

```python
def RMSE(y_true, y_pred):
    
    return np.sqrt(mean_squared_error(y_true, y_pred))

def train_predict():
    
    def rmse(y_true, y_pred):
        return K.sqrt(mse(y_true, y_pred))
    
    model = create_model()
    
    model.compile(
        loss=rmse, optimizer=keras.optimizers.Adam(learning_rate=0.001)
    )
    
    for _ in range(6):
        model.fit(X_train, y_train, validation_data=(X_val, y_val), batch_size=32, epochs=75)
    
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    print(f"Train prediction performance: {RMSE(y_train, train_pred)}")
    print(f"Test prediction performance: {RMSE(y_test, test_pred)}")
    
    predictions = model.predict(test)
    predictions = predictions.astype(int)
    
    return predictions
```

```python
pred_soup = []

for _ in range(10):
    predictions = train_predict()
    pred_soup.append(predictions)

final_predictions = np.zeros(shape=(200, 1))

for pred in pred_soup:
    final_predictions = np.add(final_predictions, pred)

final_predictions = np.divide(final_predictions, len(pred_soup)).astype(int)
```
